```python
df_clean = df.drop(['Race', 'Gender', 'Age', 'NumberofRaters', 'Female_prop',
       'Male_prop', 'Asian_prop', 'Black_prop', 'Latino_prop', 'Multi_prop',
       'Other_prop', 'White_prop', 'Afraid', 'Angry', 'Babyface',
       'Disgusted', 'Dominant', 'Feminine', 'Happy', 'Masculine', 'Prototypic',
       'Sad', 'Suitability', 'Surprised', 'Threatening', 'Trustworthy',
       'Unusual'], axis=1)
```

```python
all_images = defaultdict(list)
for filename, rating in df_clean[['Target', 'Attractive']].values:
    all_images[filename].append(rating)
```

```python
data = {}
for filename, rating in all_images.items():
    data[filename] = np.mean(rating)
```

```python
print(len(data))
ratings = dict(data)
```

```python
images = []
ys = []
path2 = path + 'Images/'
for filename in os.listdir(path2):
    img_path = path2 + filename
    img = image.load_img(img_path)
    img.thumbnail((300,300), Image.ANTIALIAS)
    img.save(path + 'Images_resized/' + filename, "JPEG")
    img = image.img_to_array(img)
    images.append(img)
    ys.append(ratings[filename[:6]])
```
